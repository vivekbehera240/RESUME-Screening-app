```python
#Cleaning the data ie urls,hashtags,mentions,special letters,punctuations
```

```python
import re
def cleanResume(txt):
    cleanTxt = re.sub(r'http\S+\s*', '', txt)  # Remove URLs (with or without trailing space)
    cleanTxt = re.sub(r'@\S+', '', cleanTxt)   # Remove mentions
    cleanTxt = re.sub(r'#\S+', '', cleanTxt)   # Remove hashtags
    cleanTxt = re.sub(r'RT|cc', '', cleanTxt)  
    cleanTxt = re.sub(r'@\S+', '', cleanTxt)
    cleanTxt = re.sub(r'[^a-zA-Z0-9\s]', '', cleanTxt)
    #cleanTxt = re.sub(r'[%s]'%re.escape("""|"#$%&()"+,-,/,*/:;<=>?@[\]^_`{|}~"""),'',cleanTxt)
    
    return cleanTxt
```

```python
#removing all the https
cleanResume("my  ###noor sayed website link is this : https://helloworld httpnoor @gmain.com and acess it ### $%^")
```

```python
df['Resume']=df['Resume'].apply(lambda x:cleanResume(x))
```

#vectorisation


```python
from sklearn.feature_extraction.text import TfidfVectorizer  # Corrected class name with proper capitalization
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requiredtext = tfidf.transform(df['Resume'])
```

```python
requiredtext
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split 
```

```python
 X_train, X_test, y_train, y_test = train_test_split( requiredtext,df['Category'] , test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
##K naeves classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier #multiclss classsifuer 
from sklearn.metrics import accuracy_score 

clf=OneVsRestClassifier(KNeighborsClassifier())
clf=KNeighborsClassifier()
clf.fit(X_train,y_train)
ypred=clf.predict(X_test)
print(accuracy_score(y_test,ypred))
```
